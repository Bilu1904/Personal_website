```python
USEFUL GENERAL CODES


==== Check the status of your repository ====

git status -> Check the status of your files in the working directory and staging area.

git log -> View the commit history for the repository.


==== Add changes and commit ====

git add . -> Stage all changes (new, modified, deleted files) for the next commit.
    
git add (file name) -> Stage a specific file for the next commit.


===== Commit changes =====

git commit -m "Your commit message" -> Commit the staged changes with a descriptive message.

git commit --amend -> Modify the most recent commit (useful for correcting commit messages or adding forgotten changes).

git resset --soft HEAD~1 -> Undo the last commit but keep changes staged for the next commit.

git reset --hard HEAD~1 -> Undo the last commit and discard changes.


==== Git Pull and Push ====

git pull -> Fetch and merge changes from the remote repository to your local repository.

git pull origin (branches name) -> Fetch and merge changes from the remote main branch to your local repository.

git push origin (branches name) -> Push your local commits to the remote repository.

git remote -v -> View the remote repositories associated with your local repository.

git clone (repository URL) -> Create a local copy of a remote repository.


==== Branching and Merging ====

git checkout (branch name) -> Switch to a different branch in your repository.

git branch -> List all branches in the repository and highlight the current branch.

git branch (new branch name) -> Create a new branch.

git merge (branch name) -> Merge the specified branch into the current branch.


==== Stashing and Discarding Changes ====

git stash -> Temporarily save changes that are not ready to be committed.

git checkout -- (file name) -> Discard changes in a specific file and revert to the last committed state.


==== Git User Configurations ====

git config --global user.name "GitHub Name"
git config --global user.email "GitHub Email"

git config --global --list

git config --list
```
